```python
from scipy.optimize import minimize
import numpy as np
def decompose_all(R, T):
    n = R.shape[1]
    x0 = np.ones(n) / n
    constraints = {'type': 'eq', 'fun': lambda x: x.sum() - 1}
    bounds = [(0, None)] * n

    results = []
    for b in T:
        res = minimize(
            lambda x: np.sum((R @ x - b)**2),
            x0=x0,
            method='SLSQP',
            bounds=bounds,
            constraints=constraints
        )
        results.append(res.x)
    return np.array(results)
```

```python
res = decompose_all(sig.values.astype(float), bulk.T.values.astype(float))
```

```python
props_df = pd.DataFrame(res)
props_df.columns = sig.columns
props_df.index = bulk.columns
props_df.to_csv("/gpfs/commons/groups/compbio/projects/rf_projects/deconv_data/deconvBench/pbmc/c2s/2b/hao800/fino_c2s_2bre_lsqclaude.csv")
```
