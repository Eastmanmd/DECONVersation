## Create Pseudobulk

```python
def create_proportional_pseudobulks(
    adata: ad.AnnData,
    cell_type_col: str, # Cell type column for pseudobulking
    n_pseudobulks: int = 1000, # Number of pseudobulk samples to be created
    target_proportion_min: float = 0.1, # Minimum target porportion
    target_proportion_max: float = 0.80, # Maximum target proportion
    n_cells_per_pseudobulk: int = 600, # number of cells used in making pseudobulk
    random_state: int = None, # random state for reprodubiility 
    min_cells_threshold: float = 0.1
):
    """
    Create pseudobulk samples with controlled cell type proportions.
    
    Parameters
    ----------
    adata : ad.AnnData
        Annotated data matrix
    cell_type_col : str
        Column name in adata.obs containing cell type labels
    n_pseudobulks : int
        Number of pseudobulk samples to generate
    target_proportion_min : float
        Minimum proportion for target cell type
    target_proportion_max : float
        Maximum proportion for target cell type
    n_cells_per_pseudobulk : int
        Number of cells to sample per pseudobulk
    random_state : int
        Random seed for reproducibility
    min_cells_threshold : float
        Warn if available cells < this fraction of needed cells
    
    Returns
    -------
    pseudobulk_df : pd.DataFrame
        Expression matrix (pseudobulks × genes)
    proportions_df : pd.DataFrame
        Cell type proportions (pseudobulks × cell types)
    """
    # Set random state for reproducibility
    rng = check_random_state(random_state)
    
    # Get unique cell types
    all_cell_types = adata.obs[cell_type_col].unique()
    
    # Output empty dataframe if no cells 
    if len(all_cell_types) == 0:
        warnings.warn("No cell types found; returning empty DataFrames.")
        return pd.DataFrame(), pd.DataFrame()
    
    # Group cell tyoes
    grouped_cells = {k: v.index.tolist() for k, v in adata.obs.groupby(cell_type_col, observed=True)}

    # Initialize empty lists to store results
    pseudobulk_data_list, proportions_data_list, sample_names = [], [], []
    
    # Generate random target proportions
    target_proportions = rng.uniform(
        low=target_proportion_min, 
        high=target_proportion_max, 
        size=n_pseudobulks
    )
    
    for i in range(n_pseudobulks):
        
        # Randomly sample cell types
        target_type = rng.choice(all_cell_types)

        # Select target proportion
        current_prop = target_proportions[i]
        
        # Calculate number of target and other cells needed
        n_target = int(np.round(n_cells_per_pseudobulk * current_prop))
        n_target = max(0, min(n_target, n_cells_per_pseudobulk))
        n_other = n_cells_per_pseudobulk - n_target # number of other cells needed to complete proportion

        # list to store selected cell indices
        selected_indices = []
        
        # Sample target cells
        if target_type in grouped_cells and n_target > 0: # if zero was drawn as a proportion
            
            available_target = len(grouped_cells[target_type]) # Number of cells 
            
            # Warning for small cell populations (# Only used if the number of target cells is greater than number of available cells to be sampled
            if available_target < n_target * min_cells_threshold:
                warnings.warn(
                    f"Pseudobulk {i+1}: Cell type '{target_type}' has only "
                    f"{available_target} cells but needs {n_target}. "
                    f"Using replacement sampling."
                )
            
                target_cells = rng.choice(
                    grouped_cells[target_type], 
                    size=n_target, 
                    replace=True
                ).tolist()
                selected_indices.extend(target_cells)
                print("oversampling as not a lot of available cells")

            # No need for replacement if enough cells are available to be sample
            else:
                target_cells = rng.choice(
                    grouped_cells[target_type], 
                    size=n_target, 
                    replace=False
                    ).tolist()
                
                selected_indices.extend(target_cells)
                #if target_type == "Epithelial":
                    #print(n_target)
                    #print(target_type)
        
        # Sample other cells
        if n_other > 0:
            other_types = [ct for ct in all_cell_types if ct != target_type] # don't include target cell type

            # Get the indices of all the other cell types not including the target cell type
            all_other_indices = sum((grouped_cells.get(ct, []) for ct in other_types), [])
            
            # Handle case where no other cells are available
            if len(all_other_indices) == 0:
                warnings.warn(
                    f"Pseudobulk {i+1}: No other cell types available. "
                    f"Filling with additional '{target_type}' cells."
                )
                if target_type in grouped_cells:
                    other_cells = rng.choice(
                        grouped_cells[target_type], 
                        size=n_other, 
                        replace=False
                    ).tolist()
                    selected_indices.extend(other_cells)
                    #print("filling with target type")
            else:
                other_cells = rng.choice(
                    all_other_indices, 
                    size=n_other, 
                    replace=False
                ).tolist()
                selected_indices.extend(other_cells)
                #print("no target types used")
        
        # Skip if no cells were selected
        if not selected_indices:
            warnings.warn(f"Pseudobulk {i+1}: No cells selected, skipping.")
            continue
        
        # Sum expression
        pb_matrix = adata[selected_indices, :].X
        pb_sum = pb_matrix.sum(axis=0).A1 if sp.issparse(pb_matrix) else pb_matrix.sum(axis=0)
        pseudobulk_data_list.append(pb_sum)
        
        # Compute proportions that match the expression aggregation
        # Count occurrences including duplicates from replacement sampling
        selected_types = adata.obs.loc[selected_indices, cell_type_col]
        type_counts = selected_types.value_counts()
        total_cells = len(selected_indices)

        #print(total_cells)
        
        # Create proportion dictionary with all cell types
        props = {ct: type_counts.get(ct, 0) / total_cells for ct in all_cell_types}
        proportions_data_list.append(props)
        
        sample_names.append(f"{target_type}_PB{i+1}")
    
    # Handle case where no pseudobulks were created
    if not pseudobulk_data_list:
        warnings.warn("No pseudobulks were created; returning empty DataFrames.")
        return pd.DataFrame(), pd.DataFrame()
    
    # Create output DataFrames
    pseudobulk_df = pd.DataFrame(
        pseudobulk_data_list, 
        index=sample_names, 
        columns=adata.var_names
    )
    proportions_df = pd.DataFrame(proportions_data_list, index=sample_names)
    
    return pseudobulk_df, proportions_df
```

```python
all_pseudobulks = []
all_cell_props = []

id_list = adata_sub.obs['id'].unique()
#id_list = ["B12-CZI041N"]

for sample_id in id_list:
    adata_subset = adata_sub[adata_sub.obs["id"] == sample_id].copy()
    
    if adata_subset.n_obs == 0:
        print(f"No cells found for ID: {sample_id}")
        continue

    pseudobulk_czi_temp, pseudobulk_czi_cell_prop_temp = create_proportional_pseudobulks(
        adata=adata_subset,
        cell_type_col="heca_lineage",
        target_proportion_min = 0.40,
        target_proportion_max = 0.85,
        n_cells_per_pseudobulk=600, 
        n_pseudobulks = 400,
        random_state=None)

    pseudobulk_czi_temp.index = [f"{sample_id}_PB_{i+1}" for i in range(0, len(pseudobulk_czi_temp))]
    pseudobulk_czi_cell_prop_temp.index = [f"{sample_id}_PB_{i+1}" for i in range(0, len(pseudobulk_czi_temp))]

    all_pseudobulks.append(pseudobulk_czi_temp)
    all_cell_props.append(pseudobulk_czi_cell_prop_temp)

# Concatenate all results
pseudobulk_czi = pd.concat(all_pseudobulks, axis=0, ignore_index=False)
pseudobulk_czi_cell_prop = pd.concat(all_cell_props, axis=0, ignore_index=False)
```

```python
def create_signature_matrix(adata, groupby):
    """
    Compute average expression per gene per cell group.
    """
    # Convert counts to dense DataFrame if sparse
    if hasattr(adata.X, "toarray"):
        expr = pd.DataFrame(adata.X.toarray(), index=adata.obs_names, columns=adata.var_names)
    else:
        expr = pd.DataFrame(adata.X, index=adata.obs_names, columns=adata.var_names)

    # Add grouping column
    expr[groupby] = adata.obs[groupby].values

    # Average expression per group (genes × groups)
    signature = expr.groupby(groupby).mean().T
    return signature
```

```python
from scipy.stats import pearsonr

def plot_true_vs_predicted(y_test_df, predicted_proportions, n_cols=6,
                           figsize_per_plot=(3, 3), dot_size=12, save_path=""):

    cols = predicted_proportions.columns
    n_rows = int(np.ceil(len(cols) / n_cols))

    fig, axes = plt.subplots(
        n_rows, n_cols,
        figsize=(n_cols * figsize_per_plot[0], n_rows * figsize_per_plot[1])
    )
    axes = axes.ravel()

    for ax, col in zip(axes, cols):
        x, y = y_test_df[col], predicted_proportions[col]

        sns.scatterplot(x=x, y=y, alpha=0.7, s=dot_size, ax=ax)

        lo, hi = min(x.min(), y.min()), max(x.max(), y.max())
        ax.plot([lo, hi], [lo, hi], "r--")

        corr, _ = pearsonr(x, y)
        ax.text(0.05, 0.95, f"r = {corr:.2f}",
                transform=ax.transAxes, va="top", fontsize=9,
                bbox=dict(boxstyle="round", facecolor="white", alpha=0.5))

        ax.set(title=col, xlabel="True", ylabel="Predicted")

    # remove unused axes
    for ax in axes[len(cols):]:
        fig.delaxes(ax)

    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=600, bbox_inches="tight")
    plt.show()
```

```python
# ============================================
# Required Imports
# ============================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr


def plot_true_vs_predicted(
    y_true_df: pd.DataFrame,
    y_pred_df: pd.DataFrame,
    n_cols: int = 3,
    figsize_per_plot: tuple = (3, 3),
    dot_size: int = 12,
    stratify_by_celltype: bool = True,
    method: str = "pearson",
    save_path: str | None = None,
):
    """
    Plot true vs predicted proportions.

    Parameters
    ----------
    y_true_df : pd.DataFrame
        Ground truth proportions (samples × cell types).

    y_pred_df : pd.DataFrame
        Predicted proportions (samples × cell types).

    n_cols : int, default=6
        Number of columns in subplot grid (only used if stratify_by_celltype=True).

    figsize_per_plot : tuple, default=(3,3)
        Size (width, height) per subplot.

    dot_size : int, default=12
        Size of scatter points.

    stratify_by_celltype : bool, default=True
        If True → one subplot per cell type.
        If False → single aggregated plot (all values combined).

    method : str, default="pearson"
        Correlation method:
            - "pearson"
            - "spearman"

    save_path : str, optional
        If provided, saves figure to this path.

    Returns
    -------
    None
    """

    if method not in ["pearson", "spearman"]:
        raise ValueError("method must be 'pearson' or 'spearman'.")

    # Align samples and cell types
    common_samples = y_true_df.index.intersection(y_pred_df.index)
    common_cols = y_true_df.columns.intersection(y_pred_df.columns)

    y_true_df = y_true_df.loc[common_samples, common_cols]
    y_pred_df = y_pred_df.loc[common_samples, common_cols]

    # ==================================================
    # Case 1: Single aggregated plot
    # ==================================================
    if not stratify_by_celltype:

        x = y_true_df.values.flatten()
        y = y_pred_df.values.flatten()

        plt.figure(figsize=(figsize_per_plot[0]*1.5,
                            figsize_per_plot[1]*1.5))

        sns.scatterplot(x=x, y=y, alpha=0.7, s=dot_size)

        lo, hi = min(x.min(), y.min()), max(x.max(), y.max())
        plt.plot([lo, hi], [lo, hi], "r--")

        if method == "pearson":
            corr, _ = pearsonr(x, y)
        else:
            corr, _ = spearmanr(x, y)

        plt.text(
            0.05, 0.95,
            f"{method.capitalize()} r = {corr:.2f}",
            transform=plt.gca().transAxes,
            va="top",
            fontsize=10,
            bbox=dict(boxstyle="round", facecolor="white", alpha=0.6),
        )

        plt.xlabel("True")
        plt.ylabel("Predicted")
        plt.title("True vs Predicted (All Cell Types)")

        plt.tight_layout()
        if save_path:
            plt.savefig(save_path, dpi=600, bbox_inches="tight")

        plt.show()
        return

    # ==================================================
    # Case 2: Stratified by cell type
    # ==================================================
    cols = common_cols
    n_rows = int(np.ceil(len(cols) / n_cols))

    fig, axes = plt.subplots(
        n_rows,
        n_cols,
        figsize=(n_cols * figsize_per_plot[0],
                 n_rows * figsize_per_plot[1]),
    )

    axes = np.array(axes).ravel()

    for ax, col in zip(axes, cols):

        x = y_true_df[col]
        y = y_pred_df[col]

        sns.scatterplot(x=x, y=y, alpha=0.7, s=dot_size, ax=ax)

        lo, hi = min(x.min(), y.min()), max(x.max(), y.max())
        ax.plot([lo, hi], [lo, hi], "r--")

        if method == "pearson":
            corr, _ = pearsonr(x, y)
        else:
            corr, _ = spearmanr(x, y)

        ax.text(
            0.05, 0.95,
            f"r = {corr:.2f}",
            transform=ax.transAxes,
            va="top",
            fontsize=9,
            bbox=dict(boxstyle="round", facecolor="white", alpha=0.6),
        )

        ax.set(title=col, xlabel="True", ylabel="Predicted")

    # Remove unused axes
    for ax in axes[len(cols):]:
        fig.delaxes(ax)

    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=600, bbox_inches="tight")

    plt.show()
```

```python
# ============================================
# Plot RMSE vs Correlation Per Cell Type
# (Correlation on X-axis, RMSE on Y-axis)
# ============================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr


def plot_rmse_vs_corr_by_class(
    true_df: pd.DataFrame,
    pred_df: pd.DataFrame,
    method: str = "pearson",
    dot_size: int = 40,
    annotate: bool = True,
    save_path: str | None = None,
):
    """
    Compute RMSE and correlation per cell type (class),
    then plot correlation (x-axis) vs RMSE (y-axis).

    Parameters
    ----------
    true_df : pd.DataFrame
        Ground truth proportions (samples × cell types)

    pred_df : pd.DataFrame
        Predicted proportions (samples × cell types)

    method : str, default="pearson"
        Correlation method:
            - "pearson"
            - "spearman"

    dot_size : int, default=60
        Size of scatter points.

    annotate : bool, default=True
        Whether to label points with class names.

    save_path : str, optional
        If provided, saves figure to this path.

    Returns
    -------
    pd.DataFrame
        DataFrame containing RMSE and correlation per class.
    """

    if method not in ["pearson", "spearman"]:
        raise ValueError("method must be 'pearson' or 'spearman'.")

    # Align samples and cell types
    common_samples = true_df.index.intersection(pred_df.index)
    common_cols = true_df.columns.intersection(pred_df.columns)

    if len(common_samples) == 0 or len(common_cols) == 0:
        raise ValueError("No overlapping samples or cell types found.")

    true_aligned = true_df.loc[common_samples, common_cols]
    pred_aligned = pred_df.loc[common_samples, common_cols]

    results = []

    for col in common_cols:
        x = true_aligned[col].values
        y = pred_aligned[col].values

        # RMSE
        rmse = np.sqrt(np.mean((x - y) ** 2))

        # Correlation
        if method == "pearson":
            corr, _ = pearsonr(x, y)
        else:
            corr, _ = spearmanr(x, y)

        results.append((col, rmse, corr))

    metrics_df = pd.DataFrame(
        results,
        columns=["cell_type", "rmse", "correlation"]
    ).set_index("cell_type")

    # Plot
    plt.figure(figsize=(6, 5))
    sns.scatterplot(
        data=metrics_df,
        x="correlation",
        y="rmse",
        s=dot_size
    )

    if annotate:
        for cell_type, row in metrics_df.iterrows():
            plt.text(
                row["correlation"],
                row["rmse"],
                cell_type,
                fontsize=14,
                ha="left",
                va="bottom"
            )

    plt.xlabel(f"Cor {}", fontsize = 18)
    plt.ylabel("RMSE", fontsize = 18)
    plt.title(" ")

    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)

    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=600, bbox_inches="tight")

    plt.show()

    return metrics_df
```

```python
plot_rmse_vs_corr_by_class(
    true_df = metrics_gf,
    pred_df= ground_truth,
    method="pearson"
)
```

```python
import os
import numpy as np
import pandas as pd


def rmse_per_file(y_test_df, predicted_df):
    """
    Compute RMSE per cell type + mean RMSE.
    Returns a Series.
    """
    common_cols = y_test_df.columns.intersection(predicted_df.columns)

    rmses = {}

    for col in common_cols:
        x = y_test_df[col].values
        y = predicted_df[col].values

        mask = ~np.isnan(x) & ~np.isnan(y)
        x = x[mask]
        y = y[mask]

        rmse = np.sqrt(np.mean((x - y) ** 2))
        rmses[col] = rmse

    rmses["mean_rmse"] = np.mean(list(rmses.values()))

    return pd.Series(rmses)


def evaluate_folder(folder_path, y_test_df, output_path=None):

    all_results = []

    for file in os.listdir(folder_path):
        if not file.endswith(".csv"):
            continue

        file_path = os.path.join(folder_path, file)

        predicted_df = pd.read_csv(file_path, index_col=0)

        # Get last three tokens from filename (without .csv)
        base = os.path.splitext(file)[0]
        tokens = base.split("_")
        row_name = "_".join(tokens[-3:])

        rmse_series = rmse_per_file(y_test_df, predicted_df)
        rmse_series.name = row_name

        all_results.append(rmse_series)

    results_df = pd.DataFrame(all_results)

    if output_path:
        results_df.to_csv(output_path)

    return results_df
```

```python
results = evaluate_folder(
    folder_path="/gpfs/commons/groups/compbio/projects/ao_projects/ml_deconv_data/pseudo_bulk/emb_mode_files/results/",
    y_test_df=ground_truth,
    output_path=None
)
```

```python
results_df = results.sort_values(by="mean_rmse", ascending=True)
```
