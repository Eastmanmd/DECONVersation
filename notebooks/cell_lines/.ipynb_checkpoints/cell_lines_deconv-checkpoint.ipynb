```python
def plot_pca(df, color_labels=None, n_components=2, title="PCA Plot", figsize=(10, 8), save_path = "../plots/temp.png"):
    """
    Plot PCA from a dataframe (rows = samples, columns = features).

    Parameters
    ----------
    df : pd.DataFrame
        Input data with numeric variables as columns.
    color_labels : array-like, optional
        Labels or continuous values for coloring points.
    n_components : int, default=2
        Number of principal components to compute.
    title : str, default="PCA Plot"
        Plot title.
    """

    # Drop non-numeric columns if any
    df_numeric = df.select_dtypes(include=[np.number])
    if df_numeric.empty:
        raise ValueError("No numeric columns found for PCA.")

    # Standardize data
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(df_numeric)

    # Fit PCA
    pca = PCA(n_components=n_components, random_state=42)
    pcs = pca.fit_transform(scaled_data)

    # Create DataFrame for plotting
    pca_df = pd.DataFrame(pcs[:, :2], columns=["PC1", "PC2"], index=df.index)
    if color_labels is not None:
        pca_df["Color"] = color_labels

    plt.figure(figsize=figsize)
    
    # Determine colormap type
    if color_labels is not None:
        if pd.api.types.is_numeric_dtype(pd.Series(color_labels)):
            # Continuous variable
            scatter = plt.scatter(
                pca_df["PC1"], pca_df["PC2"],
                c=pca_df["Color"], cmap="viridis", s=40, alpha=0.8, edgecolors="k"
            )
            plt.colorbar(scatter, label="Color variable")
        else:
            # Categorical variable
            sns.scatterplot(
                data=pca_df, x="PC1", y="PC2", hue="Color",
                palette="Set2", s=40, alpha=0.8, edgecolor="k"
            )
    else:
        # No color variable
        plt.scatter(pca_df["PC1"], pca_df["PC2"], color="gray", s=40, alpha=0.8, edgecolors="k")

    # Add explained variance info
    var_exp = pca.explained_variance_ratio_ * 100
    plt.xlabel(f"PC1 ({var_exp[0]:.2f}%)",fontsize=16)
    plt.ylabel(f"PC2 ({var_exp[1]:.2f}%)", fontsize=16)
    plt.legend().remove()
    plt.title(title)
    #plt.savefig(save_path, dpi=600, bbox_inches='tight', transparent=False)
    plt.tight_layout()
    plt.show()
```

```python
def create_signature_matrix(adata, groupby):
    """
    Compute average expression per gene per cell group.
    """
    # Convert counts to dense DataFrame if sparse
    if hasattr(adata.X, "toarray"):
        expr = pd.DataFrame(adata.X.toarray(), index=adata.obs_names, columns=adata.var_names)
    else:
        expr = pd.DataFrame(adata.X, index=adata.obs_names, columns=adata.var_names)

    # Add grouping column
    expr[groupby] = adata.obs[groupby].values

    # Average expression per group (genes × groups)
    signature = expr.groupby(groupby).mean().T
    return signature
```

```python
def plot_true_vs_predicted(y_test_df, predicted_proportions, n_cols=6, figsize_per_plot=(3, 3), dot_size=5, save_path=""):

    # Get column names (cell types or lineages)
    cols = predicted_proportions.columns
    n_rows = int(np.ceil(len(cols) / n_cols))

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * figsize_per_plot[0], n_rows * figsize_per_plot[1]))
    axes = axes.flatten()

    for i, col in enumerate(cols):
        ax = axes[i]
        x = y_test_df[col]
        y = predicted_proportions[col]

        # Scatterplot
        sns.scatterplot(x=x, y=y, alpha=0.7, ax=ax, s=dot_size)

        # Identity line
        min_val = min(x.min(), y.min())
        max_val = max(x.max(), y.max())
        ax.plot([min_val, max_val], [min_val, max_val], 'r--')

        # Correlation coefficient
        corr, _ = pearsonr(x, y)
        ax.text(
            0.05, 0.95, f"r = {corr:.2f}",
            transform=ax.transAxes,
            verticalalignment='top',
            fontsize=9,
            bbox=dict(boxstyle="round", facecolor="white", alpha=0.5)
        )

        # Titles and labels
        ax.set_title(f'{col}')
        ax.set_xlabel('True')
        ax.set_ylabel('Predicted')

    # Hide any unused subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.savefig(save_path, dpi=600, bbox_inches='tight', transparent=False)
    plt.tight_layout()
    plt.show()
```

```python
import os
import glob
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
from sklearn.metrics import mean_squared_error

def summarize_proportion_predictions(
    results_dir,
    true_file,
    pattern="_pred.csv",
    output_summary=None
):
    # --- Step 1: find matching files ---
    pred_files = glob.glob(os.path.join(results_dir, f"*{pattern}"))
    if not pred_files:
        raise FileNotFoundError(f"No files found in {results_dir} matching pattern '{pattern}'")

    # --- Step 2: read true proportions ---
    true_df = pd.read_csv(true_file, index_col=0)
    true_df = true_df.loc[:, ~true_df.columns.duplicated()]  # ensure no duplicate columns

    summary_records = []

    # --- Step 3: loop over each prediction file ---
    for pred_path in pred_files:
        pred_df = pd.read_csv(pred_path, index_col=0)
        
        # Align columns with true proportions
        common_cols = [col for col in pred_df.columns if col in true_df.columns]
        if not common_cols:
            print(f"⚠️ No matching columns in {os.path.basename(pred_path)} — skipped.")
            continue
        
        pred_df = pred_df[common_cols]
        true_sub = true_df[common_cols].loc[pred_df.index]

        # --- Step 4: calculate metrics per cell type ---
        for cell_type in common_cols:
            y_true = true_sub[cell_type]
            y_pred = pred_df[cell_type]

            corr, _ = pearsonr(y_true, y_pred)
            rmse = np.sqrt(mean_squared_error(y_true, y_pred))

            summary_records.append({
                "file": os.path.basename(pred_path),
                "cell_type": cell_type,
                "correlation": corr,
                "rmse": rmse
            })

    summary_df = pd.DataFrame(summary_records)

    # --- Step 5: optional — overall summary ---
    overall_summary = (
        summary_df.groupby("file")[["correlation", "rmse"]]
        .mean()
        .reset_index()
        .rename(columns={"correlation": "mean_correlation", "rmse": "mean_rmse"})
    )

    # --- Step 6: save or return ---
    if output_summary:
        summary_df.to_csv(output_summary, index=False)
        print(f"✅ Summary saved to: {output_summary}")

    return summary_df, overall_summary
```

```python
summary_df, overall_summary = summarize_proportion_predictions("/gpfs/commons/groups/compbio/projects/ao_projects/ml_deconv_data_cell_lines/metrics_nnls/",
                                                               "/gpfs/commons/groups/compbio/projects/ao_projects/ml_deconv_data/cell_lines/test/cell_proportions.csv",
                                                               pattern="_nnls_cell_proportions.csv",
    output_summary=None)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# --- Aggregate across cell types (mean RMSE per method) ---
method_summary = (
    overall_summary.groupby("method", as_index=False)["mean_rmse"]
    .mean()
    .sort_values("mean_rmse", ascending=True)
)

# --- Define fixed colors for each method (reuse consistently) ---
method_colors = {
    "MuSiC": "#808080",         # grey
    "Geneformer": "#1f77b4", # blue
    "C2S": "#aec7e8", # light blue
    "counts": "#2ca02c", # green
    "pca": "#98df8a", # light green
    "scVI": "#d62728",        # red
    "counts": "#ff9896",      # light red
}

# Match colors to method order in plot
palette = [method_colors.get(m, "#cccccc") for m in method_summary["method"]]

# --- Plot simple bar plot (RMSE by method) ---
plt.figure(figsize=(10, 7))
sns.barplot(
    data=method_summary,
    x="method",
    y="mean_rmse",
    order=method_summary["method"],
    palette=palette,
    edgecolor="black",
    width=0.6  # Reduce bar width (default is 0.8)
)

# --- Beautify ---
plt.title("", fontsize=14)
plt.xlabel("Method", fontsize=18)
plt.ylabel("Mean RMSE", fontsize=18)
plt.xticks(rotation=30, ha="right", fontsize=18)
plt.yticks(np.arange(0, 0.6, 0.05), fontsize=18)  # Set y-axis 0–0.5 in 0.025 steps
plt.ylim(0, 0.5)  # Ensure axis limits
sns.despine()

# --- Save and show ---
plt.tight_layout()
#plt.savefig("plots/rmse_rf.png", dpi=600, bbox_inches="tight", transparent=False)
plt.show()
```
