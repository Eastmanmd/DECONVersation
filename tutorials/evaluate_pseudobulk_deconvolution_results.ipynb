# Benchmarking scFM methods on pseudobulk data  
<b> Evaluations </b> <br>
Here we benchmark scFM-based deconvolution methods using pseudobulk data. For details on pseudobulk data generation and method outputs, see the pseudobulk tutorials. Results are stored in the format `method_estimated_proportions.csv`. We use `load_results()` to load all results into a single dataframe, then `plot_cell_type_heatmaps()` and `plot_global_comparison()` to visualize global performance. `plot_cell_type_heatmaps()` shows per-cell-type correlation and RMSE heatmaps across methods; `plot_global_comparison()` summarizes global correlation with bar plots.

```python
res_df = load_results(folder_path = '../results/',
                      ground_truth_file = '../outputs/pseudobulk_true_proportions.csv')

# Drop ground truth results 
res_df = res_df[res_df.Model != "pseudobulk"]
```

```python
# Remove global correlations/RMSEs
res_df_cell_type = res_df[res_df['CellType'] != 'Global'].copy()
res_df_global = res_df[res_df['CellType'] == 'Global'].copy()

# Plot
plot_cell_type_heatmaps(res_df_cell_type)
plot_global_comparison(res_df_global)
```
